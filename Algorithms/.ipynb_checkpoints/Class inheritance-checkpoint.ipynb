```python
""" 
Формат входных данных
В первой строке входных данных содержится целое число n - число классов.

В следующих n строках содержится описание наследования классов. В i-й строке указано от каких классов наследуется i-й класс. 
Класс может ни от кого не наследоваться. Гарантируется, что класс не наследуется сам от себя (прямо или косвенно), что класс не наследуется явно от одного класса более одного раза.

В следующей строке содержится число q - количество запросов.

В следующих q строках содержится описание запросов в формате <имя класса 1> <имя класса 2>.
Имя класса – строка, состоящая из символов латинского алфавита, длины не более 50.

Формат выходных данных
Для каждого запроса выведите в отдельной строке слово "Yes", если класс 1 является предком класса 2, и "No", если не является.
"""

Sample Input:

4
A
B : A
C : A
D : B C
4
A B
B D
C D
D A

Sample Output:

Yes
Yes
Yes
No
```

```python
def rec(parent, child):
    if parent in namespace and child in namespace and parent == child:
        return "Yes"
    if (parent not in namespace) or (child not in namespace) or (namespace[parent][1] == []) or (namespace[child][0] == []):
        return "No"
    else:
        if child in namespace[parent][1]:
            return "Yes"
        else:
            for par in namespace[child][0]:
                if par == parent:
                    return "Yes"
                    break
                else: 
                    if rec(parent, par) == "No":
                        continue
                    else:
                        return "Yes"
                        break
            return "No"
        
namespace = {} # пространство имен ведем списком, где первый элемент - список наследников, а второй - список родителей
n = int(input())
for _ in range(n):
    x = list(map(lambda s: s.strip(), input().split(":"))) # разделяем родительский класс и классы-наследники
    x = list(map(lambda s: s.split(" "), x)) # разделяем классы наследники
    parent = x[0][0]
    if len(parent) > 50 or parent.isalpha() == False: # проверка условий на имена классов
        continue
    if len(x) > 1:
        children = x[1]
        for clas in children:
            if len(clas)>50 or clas.isalpha() == False: # проверка условий на имена классов
                continue
    if len(x) == 1: # если класс без наследников
        if parent not in namespace:
            namespace[parent] = [[], []]
        else:
            continue
    else:
        if parent not in namespace: # если не в словаре, то инициализируем с указанием наследников из ввода
            namespace[parent] = [children, []]
        else: 
            for clas in children: # иначе добавляем наследников к списку наследников в словаре
                namespace[parent].append(clas)
        for clas in children: 
            if clas not in namespace: 
                namespace[clas] = [[], [parent]] # инициализируем и наследников, если не в словаре
            else:
                namespace[clas][1].append([parent]) # иначе добавляем родителя к уже имеющимся спискам по наследникам 
m = int(input())                
for _ in range(m):
    req = list(map(lambda s: s.strip(), input().split(" "))) # считывание и обработка запросов
    print(rec(req[0], req[1]))
```

```python
""" 
Формат входных данных
В первой строке входных данных содержится целое число n - число классов исключений.

В следующих n строках содержится описание наследования классов. В i-й строке указано от каких классов наследуется i-й класс. 
Класс может ни от кого не наследоваться. Гарантируется, что класс не наследуется сам от себя (прямо или косвенно), что класс не наследуется явно от одного класса более одного раза.

В следующей строке содержится число m - количество обрабатываемых исключений.
Следующие m строк содержат имена исключений.
Гарантируется, что никакое исключение не обрабатывается дважды.

Формат выходных данных
Выведите в отдельной строке имя каждого исключения, обработку которого можно удалить из кода, не изменив при этом поведение программы. 
Имена следует выводить в том же порядке, в котором они идут во входных данных.
""" 
Пример теста 1
Рассмотрим код

try:
   foo()
except ZeroDivision :
   print("ZeroDivision")
except OSError:
   print("OSError")
except ArithmeticError:
   print("ArithmeticError")
except FileNotFoundError:
   print("FileNotFoundError")


...

Sample Input:

4
ArithmeticError
ZeroDivisionError : ArithmeticError
OSError
FileNotFoundError : OSError
4
ZeroDivisionError
OSError
ArithmeticError
FileNotFoundError

Sample Output:
FileNotFoundError
```

```python
"""
рекурсивно просматриваем всех наследников и возвращаем их список
"""
def rec(error, visited): 
    if namespace[error][1] == []:
        pass
    else:
        for child in namespace[error][1]:
            visited.append(child)
            rec(child, visited)
    return visited

namespace = {} # пространство имен ведем списком, где первый элемент - список наследников, а второй - список родителей
n = int(input())

for _ in range(n):
    x = list(map(lambda s: s.strip(), input().split(":"))) # разделяем родительский класс и классы-наследники
    x = list(map(lambda s: s.split(" "), x)) # разделяем классы наследники
    parent = x[0][0]
    if len(x) == 1:
        if parent not in namespace:
            namespace[parent] = [[], []]
        else:
            continue
    else:
        children = x[1]
        if parent not in namespace:
            namespace[parent] = [children.copy(), []]
        else: 
            for clas in children:
                namespace[parent][0].append(clas)
        for clas in children:
            if clas not in namespace:
                namespace[clas] = [[], parent.copy()]
            else:
                if clas not in namespace[clas][1]:
                    namespace[clas][1].append(parent)
m = int(input()) 
inp = []
visited = []
res = []

for _ in range(m): # считывание запросов
    inp.append(input())
    
for error in inp: 
    if error not in namespace:
        continue
    if error in visited: # если уже не нужно обрабатывать данное исключение и его еще нет в списке лишних 
        if error not in res:
            res.append(error)
    else: # если нужно обрабатывать, то для для будущих итераций добавляем исключение и его наследников в список
        visited.append(error)
        visited = rec(error, visited)

for err in res:
    print(err)
```

```python
""" 
Вам дано описание наследования классов в формате JSON.
Описание представляет из себя массив JSON-объектов, которые соответствуют классам. 
У каждого JSON-объекта есть поле name, которое содержит имя класса, и поле parents, которое содержит список имен прямых предков.
"""
Пример:
[{"name": "A", "parents": []}, {"name": "B", "parents": ["A", "C"]}, {"name": "C", "parents": ["A"]}]

Эквивалент на Python:

class A:
    pass

class B(A, C):
    pass

class C(A):
    pass

""" 
Гарантируется, что никакой класс не наследуется от себя явно или косвенно, и что никакой класс не наследуется явно от одного класса более одного раза.

Для каждого класса вычислите предком скольких классов он является и выведите эту информацию в следующем формате.

<имя класса> : <количество потомков>

Выводить классы следует в лексикографическом порядке.
"""
Sample Input:

[{"name": "A", "parents": []}, {"name": "B", "parents": ["A", "C"]}, {"name": "C", "parents": ["A"]}]
Sample Output:

A : 3
B : 1
C : 2
```

```python
def rec(clas, child):
    for key, value in namespace.items():
        if clas in value and key not in child: # если потомок просматриваемого класса, то запоминаем и рекурсивно смотрим потомков потомка
            child.append(key)
            rec(key, child)
    return len(child) + 1

import json
namespace = {}
inp = json.loads(input())

for obj in inp: # заполнение словаря пространства имен
    namespace[obj["name"]] = obj["parents"]

for key in sorted(namespace): # для каждого ключа считаем число потомков с помощью рекурсивной функции
    children = []
    print(key, ":", rec(key, children))
```
