```python
""" 
Проверка свойства дерева поиска
Проверить, является ли данное двоичное дерево деревом поиска.
Вход. Двоичное дерево.
Выход. Проверить, является ли оно корректным деревом
поиска: верно ли, что для любой вершины дерева её ключ
больше всех ключей в левом поддереве данной вершины и меньше всех ключей в правом поддереве.
"""
```

```python
import sys
sys.setrecursionlimit(100000)
class node:
    def __init__(self, key=None, left=None, right=None):
        self.key = key
        self.left = left
        self.right = right        
        
    def __str__(self):        
        return 'Key: ' + str(self.key) + ' Left: ' + str(self.left) + ' Right: ' + str(self.right)
class Tree:
    def __init__(self):
        self.nodes = []
        self.data = []

    def newNode(self, key, left=None, right=None):
        self.nodes.append(node(key, left, right))
        self.data.append(key)

    def __str__(self):        
        temp = ""
        for node in self.nodes:
            temp += str(node) + "\n"
        return temp       

def left_walk(node, tree):
    if tree.nodes[node.left].key == -1:
        return 
    if node.key > tree.nodes[node.left].key: # если родительский элемент больше левого ребенка, то продолжаем проход, иначе возвращаем -1
        left_walk(tree.nodes[node.left], tree)   
    else:
        return -1
    
def right_walk(node, tree):
    if tree.nodes[node.right].key == -1:
        return 
    if node.key<tree.nodes[node.right].key: # если родительский элемент меньше правого ребенка, то продолжаем проход, иначе возвращаем -1
        right_walk(tree.nodes[node.right], tree)   
    else:
        return -1
    
def check(node, tree, cash, act="None"):
    if node.key == -1:
        return 
    if (node.key <= tree.nodes[node.right].key or tree.nodes[node.right].key == -1) and node.key > tree.nodes[node.left].key:
        if act == "more": # через act проверяем свойство для всего левого или правого поддерева
            if node.key >= cash:
                check(tree.nodes[node.left], tree, cash, act="more")
                check(tree.nodes[node.right], tree, cash, act="more")
            else:
                return 0
        elif act == "less":
            if node.key < cash:
                check(tree.nodes[node.left], tree, cash, act="less")
                check(tree.nodes[node.right], tree, cash, act="less")
            else:
                return 0
        else: #для корня 
                check(tree.nodes[node.left], tree, cash, act="less")   
                check(tree.nodes[node.right], tree, cash, act="more")
    else:          
        return 0   
            
n = int(input())
t = Tree() 
data = []
flag = 1
for _ in range(n): # считываем значения
    temp = list(map(int, input().split()))
    data.append(temp)
data.append([-1, -1, -1]) # для создания листьев    

for el in data: # строим дерево
    t.newNode(el[0], el[1], el[2])
flag = check(t.nodes[0], t, t.nodes[0].key) # проверяем свойство
try:
    if flag == 0:
        print("INCORRECT")
    else:
        print("CORRECT")
except Exception:
    print("CORRECT") 
```

```python
# Данная задача полностью аналогична предыдущей, но проверять теперь нужно более общее свойство. Дереву разрешается содержать
# равные ключи, но они всегда должны находиться в правом поддереве.
# Формально, двоичное дерево называется деревом поиска, если для любой вершины её ключ больше всех ключей из её левого поддерева
```

```python
import sys
sys.setrecursionlimit(1000000)
class node:
    def __init__(self, key=None, left=None, right=None):
        self.key = key
        self.left = left
        self.right = right        
        
    def __str__(self):        
        return 'Key: ' + str(self.key) + ' Left: ' + str(self.left) + ' Right: ' + str(self.right)
class Tree:
    def __init__(self):
        self.nodes = []
        self.data = []

    def newNode(self, key, left=None, right=None):
        self.nodes.append(node(key, left, right))
        self.data.append(key)

    def __str__(self):        
        temp = ""
        for node in self.nodes:
            temp += str(node) + "\n"
    
def check(node, tree, mi=-2**32, ma=2**32): # для проверки запоминаем максимальное и минимальное значение по поддереву, инициализация очень маленьким и очень большим числом
    if node.key == -1:
        return 1 
    value = node.key
    left = tree.nodes[node.left].key
    right = tree.nodes[node.right].key
    if left != -1: # если не лист, то должно быть не больше родителя, иначе нарушение
        if left >= value:
            return 0
    if right != -1: # если не лист, то должно быть больше родителя, иначе нарушение
        if right < value:
            return 0        
    if value < mi or value >= ma:
        return 0
    else:
        return check(tree.nodes[node.left], tree, mi, value) and check(tree.nodes[node.right], tree, value, ma) 

n = int(input())
t = Tree() 
data = []
flag = 1
for _ in range(n): # считываем значения
    temp = list(map(int, input().split()))
    data.append(temp)
data.append([-1, -1, -1]) # для создания листьев     

for el in data: # строим дерево
    t.newNode(el[0], el[1], el[2])
flag = check(t.nodes[0], t) # проверяем свойство
if flag == 0:
    print("INCORRECT")
else:
    print("CORRECT") 
```
