```python
""" 
По данной непустой строке, состоящей из строчных букв латинского алфавита, постройте оптимальный беспрефиксный код. 
В первой строке выведите количество различных букв k, встречающихся в строке, и размер получившейся закодированной строки. 
В следующих k строках запишите коды букв в формате "letter: code". В последней строке выведите закодированную строку.
"""
Sample Input 1:

a

Sample Output 1:

1 1
a: 0
0

Sample Input 2:

abacabad

Sample Output 2:

4 14
a: 0
b: 10
c: 110
d: 111
01001100100111
```

```python
import collections 
import heapq as hq 

""" 
классы для построения беспрефиксного дерева
"""
class node:
    def __init__(self, data=None, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right        
        
    def __str__(self):        
        return 'Node ['+str(self.data)+']'
    
class Tree:
    def __init__(self):
        self.nodes = []
        self.data = []

    def newNode(self, data, left=None, right=None):
        self.nodes.append(node(data, left, right))
        self.data.append(data)

string = input()
tree = Tree()
heap = []
cod = {}
code = {}
freq = dict(collections.Counter(string)) # словарь частот символов для кодирования

for key, value in freq.items(): # на основании частот строим кучу
    hq.heappush(heap, (value,key))

for el in heap: # на основании кучи инициализируем узлы дерева
    tree.newNode(el) 

if len(heap) > 1: 
    while len(heap)>1: # объединяем узлы с самыми большими значениями в дереве и добавляем суммарное значение в кучу
        left = hq.heappop(heap)
        right = hq.heappop(heap)
        data = (left[0]+right[0], left[1]+right[1])
        tree.newNode(data, left, right)
        hq.heappush(heap, data)

def rec(char, tree, dic, node=-1, code=""): 
    if tree.data[node][1] == char:
        dic[char] = code
    if tree.nodes[node].left == None and tree.nodes[node].right == None:
        return
    rec(char, tree, dic, tree.data.index(tree.nodes[node].left), code+"0")
    rec(char, tree, dic, tree.data.index(tree.nodes[node].right), code+"1")
    
def encode (string, tree, dic):
    if len(freq) == 1:
        dic[string[0]] = "0"
        return dic
    for char in string:
        if char not in dic:
            rec(char, tree, dic)
    return dic

encode(string, tree, cod)

for key, value in sorted(cod.items(), key = lambda item: item[1]):
    code[key] = value

def codify(string, cod=""):
    for char in string:
        cod += code[char]
    return cod

print(str(len(freq)) + " " + str(len(codify(string))))
for key, value in cod.items():
    print(key + ": " + str(value))
print(codify(string))
```

```python
k, l = map(int, input().split())
dic = {}
res = ""
cash = ""

for _ in range(k):
    key, value = input().split(": ")
    dic[key] = value
    
code = input().rstrip()

for char in code:
    cash += char
    if cash in dic.values():
        res += list(dic.keys())[list(dic.values()).index(cash)]
        cash = ""
        
print(res)
```
