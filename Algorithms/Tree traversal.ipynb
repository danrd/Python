```python
"""
Обход двоичного дерева
Построить in-order, pre-order и post-order обходы данного двоичного дерева.
Вход. Двоичное дерево.
Выход. Все его вершины в трёх разных порядках: in-order,
pre-order и post-order.
"""
```

```python
import collections 
import heapq as hq 

class node:
    def __init__(self, key=None, left=None, right=None):
        self.key = key
        self.left = left
        self.right = right        
        
    def __str__(self):        
        return 'Key: ' + str(self.key) + ' Left: ' + str(self.left) + ' Right: ' + str(self.right)

class Tree:
    def __init__(self):
        self.nodes = []
        self.data = []

    def newNode(self, key, left=None, right=None):
        self.nodes.append(node(key, left, right))
        self.data.append(key)

    def __str__(self):        
        temp = ""
        for node in self.nodes:
            temp += str(node)+"\n"
        return temp       
        
def in_order(node, tree):
    if node.key == -1:
        return 
    in_order(tree.nodes[node.left], tree)
    print(node.key, end=" ")
    in_order(tree.nodes[node.right], tree)

def pre_order(node, tree):
    if node.key == -1:
        return 
    print(node.key, end=" ")
    pre_order(tree.nodes[node.left], tree)
    pre_order(tree.nodes[node.right], tree)

def post_order(node, tree):
    if node.key == -1:
        return 
    post_order(tree.nodes[node.left], tree)
    post_order(tree.nodes[node.right], tree)
    print(node.key, end=" ")

n = int(input())
t = Tree() 
data = []

for i in range(n):
    temp = list(map(int, input().split()))
    data.append(temp)
data.append([-1, -1, -1]) # для узлов без значения  

for el in data:
    t.newNode(el[0], el[1], el[2])
    
# делаем обходы из корня
in_order(t.nodes[0], t)
print()

pre_order(t.nodes[0], t)
print()

post_order(t.nodes[0], t)    
print()
```
