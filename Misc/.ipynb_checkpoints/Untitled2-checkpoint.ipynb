```python
def blackjack_hand_greater_than(hand_1, hand_2):
    """
    Return True if hand_1 beats hand_2, and False otherwise.
    
    In order for hand_1 to beat hand_2 the following must be true:
    - The total of hand_1 must not exceed 21
    - The total of hand_1 must exceed the total of hand_2 OR hand_2's total must exceed 21
    
    Hands are represented as a list of cards. Each card is represented by a string.
    
    When adding up a hand's total, cards with numbers count for that many points. Face
    cards ('J', 'Q', and 'K') are worth 10 points. 'A' can count for 1 or 11.
    
    When determining a hand's total, you should try to count aces in the way that 
    maximizes the hand's total without going over 21. e.g. the total of ['A', 'A', '9'] is 21,
    the total of ['A', 'A', '9', '3'] is 14.
    
    Examples:
    >>> blackjack_hand_greater_than(['K'], ['3', '4'])
    True
    >>> blackjack_hand_greater_than(['K'], ['10'])
    False
    >>> blackjack_hand_greater_than(['K', 'K', '2'], ['3'])
    False
    """
    royal = ["J", "Q", "K"]
    def pointify(hand):
        points = 0
        remainder = 21
        aces = 0
        for card in hand:
            if card!="A":
                if card in royal:
                    points += 10
                    remainder = remainder-10
                else:
                    points += int(card)
                    remainder = remainder-int(card)
            else:
                aces += 1
        if remainder>=11 and aces==1:
            points += 11
        else:
            points += 1
        if remainder-11-(aces-1)>=0 and aces>1:
            points += 11+aces-1
        else:
            points += aces
        return points
    def greater(hand1, hand2):
        if pointify(hand1)<=21 and pointify(hand2)<=21 and pointify(hand1)>pointify(hand2): 
            return True
        elif pointify(hand1)<=21 and pointify(hand2)>21:
            return True
        else:
            return False
    return greater(hand_1, hand_2)
```

```python
import collections 
import heapq as hq 
import sys
sys.setrecursionlimit(100000)
class node:
    def __init__(self, key = None, left = None, right = None):
        self.key = key
        self.left = left
        self.right = right        
        
    def __str__(self):        
        return 'Key: '+str(self.key)+' Left: '+str(self.left)+' Right: '+str(self.right)
class Tree:
    def __init__(self):
        self.nodes = []
        self.data = []

    def newNode(self, key, left = None, right = None):
        self.nodes.append(node(key, left, right))
        self.data.append(key)

    def __str__(self):        
        temp = ""
        for node in self.nodes:
            temp += str(node)+"\n"
        return temp       
    
def check(node, tree, cash, act="None"):
    if node.key==-1:
        return 
    if (node.key<=tree.nodes[node.right].key or tree.nodes[node.right].key==-1) and node.key>tree.nodes[node.left].key:
        if act=="more":
            if node.key>=cash:
                check(tree.nodes[node.left], tree, cash, act="more")
                check(tree.nodes[node.right], tree, cash, act="more")
            else:
                return 0
        elif act=="less":
            if node.key<cash:
                check(tree.nodes[node.left], tree, cash, act="less")
                check(tree.nodes[node.right], tree, cash, act="less")
            else:
                return 0
        else:
                check(tree.nodes[node.left], tree, cash, act="less")   
                check(tree.nodes[node.right], tree, cash, act="more")
    else:          
        return 0   
            
n = int(input())
t = Tree() 
data = []
flag = 1
for i in range(n):
    temp = list(map(int, input().split()))
    data.append(temp)
data.append([-1, -1, -1])    

for el in data:
    t.newNode(el[0], el[1], el[2])
r = t.nodes[0].key
res = check(t.nodes[0], t, r)
if res==0:
    flag = 0
if flag==1:
    print("CORRECT")
else:
    print("INCORRECT")
```
